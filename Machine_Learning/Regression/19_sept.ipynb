<h3>Train Test Split</h3>

```python
df2
```

```python
x = df.drop('price',axis='columns')
x.head()
```

```python
df2['owner'] = df2['owner'].map(owner_dictonary)
```

```python
# x and y
x = df2.drop('price',axis='columns')
y = df2[['price']]
```

```python
x
```

```python
y
```

```python
# 1.split full dataset (features and target)
# 2.split data into two parts (major part for training and minor part for training) (x_train,x_test,y_train,y_test)
# 3.training
# 4.testing
```

```python
from sklearn.model_selection import train_test_split
```

```python
## train and testing split data (into 4 parts)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
x_test.head()
```

```python
y_test.head()
```

```python
from sklearn.linear_model import LinearRegression # class
```

```python
lnr = LinearRegression()   # object/instance of LinearRegression class
```

```python
x_train.head()
```

```python
lnr.fit(x_train,y_train) # fit function for training the algo
```

```python
x_train.shape
```

```python
## model evaluation phase start
print("training score : ",lnr.score(x_train,y_train)*100)
print("testing score : ",lnr.score(x_test,y_test)*100)
```

```python
## algo is now trained and it can give predictions
prediction = lnr.predict(x_test)
```
